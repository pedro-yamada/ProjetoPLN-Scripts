# Busca de 3 resumos mais similares

````
# Definir o modelo de embedding
embedding_model = 'paraphrase-multilingual-MiniLM-L12-v2'
sentence_encoder = SentenceTransformer(embedding_model)
````

````
# Função para carregar o dicionário de descrições e embeddings
def load_dictionary(file_path: str) -> Dict[str, str]:
    with open(file_path, 'r') as file:
        metadatas = json.load(file)

    # Inicializa o dicionário de overviews
    overviews = {}

    # Itera sobre os dados para extrair os overviews
    for movie_name in metadatas.keys():
        try:
            if metadatas[movie_name]['tmdb']['overview'] != "":
                overviews[movie_name] = metadatas[movie_name]['tmdb']['overview']
        except (KeyError, TypeError) as e:
            print(f"Erro ao processar dados para '{movie_name}': {e}")

    return overviews

# Função para criar a base de dados vetorial Qdrant
def load_qdrant_client(overviews: Dict[str, str]) -> QdrantClient:
    # Inicializa o cliente Qdrant
    vector_DB_client = QdrantClient(":memory:")

    # Cria o dicionário de embeddings
    embedding_dict = {}
    for descricao, overview in overviews.items():
        try:
            # Calcula o embedding para o overview e armazena
            embedding = sentence_encoder.encode(overview)
            embedding_dict[descricao] = np.array(embedding)
            print(f"Embedding criado para '{descricao}': {embedding[:5]}...")  # Exibe uma amostra do embedding
        except Exception as e:
            print(f"Erro ao criar embedding para '{descricao}': {e}")

    # Verificar se o dicionário de embeddings não está vazio
    if not embedding_dict:
        raise ValueError("O dicionário de embeddings está vazio. Verifique o conteúdo do dicionário de overviews.")

    # Obtém a dimensão dos embeddings
    try:
        first_embedding = next(iter(embedding_dict.values()))
        embedding_dim = first_embedding.shape[0]
    except StopIteration:
        raise ValueError("Não foi possível obter a dimensão dos embeddings. O dicionário de embeddings pode estar vazio.")

    # Cria a coleção de filmes na base
    vector_DB_client.create_collection(
        collection_name="MOVIES",
        vectors_config=models.VectorParams(
            size=embedding_dim,
            distance=models.Distance.COSINE
        ),
    )

    # Insere os pontos na base
    vector_DB_client.upload_points(
        collection_name="MOVIES",
        points=[
            models.PointStruct(
                id=idx,
                vector=embedding_dict[descricao].tolist(),
                payload={"overview": overview}
            )
            for idx, (descricao, overview) in enumerate(overviews.items())
        ],
    )

    return vector_DB_client

# Função para recuperar filmes relevantes
def retrieve_relevant_movies(
    embedding_model: SentenceTransformer,
    vector_DB_client: QdrantClient,
    query: str,
    num_to_retrieve: int
) -> List[Dict[str, Any]]:
    """
    Retorna filmes similares à consulta usando o encoder de sentenças e Qdrant.
    """
    # Cria o embedding da consulta
    query_vector = embedding_model.encode(query).tolist()

    # Calcula a distância e busca os filmes similares
    hits = vector_DB_client.search(
        collection_name="MOVIES",
        query_vector=query_vector,
        limit=num_to_retrieve,
    )

    # Extrai os resultados relevantes
    relevant_movies = []
    for hit in hits:
        movie_id = hit.id
        payload = hit.payload
        movie_name = [name for name, overview in overviews.items() if overview == payload['overview']]
        relevant_movies.append({
            'movie_name': movie_name[0] if movie_name else 'Unknown',
            'overview': payload['overview']
        })

    return relevant_movies
````

````
# Carregar o dicionário de descrições e embeddings
file_path = 'Movie-Script-Database/scripts/metadata/clean_files_meta.json'
overviews = load_dictionary(file_path)
# Exibir exemplos de overviews
print("Exemplos de overviews:")
for movie, overview in list(overviews.items())[:5]:  # Exibir apenas os primeiros 5 exemplos
    print(f"Filme: {movie}\nOverview: {overview}\n")
````

````
# Carregar o encoder de textos
model_name = 'paraphrase-multilingual-MiniLM-L12-v2'
sentence_encoder = SentenceTransformer(model_name)

# Carregar a base de dados Qdrant
vector_DB_client = load_qdrant_client(overviews)
````

### Definindo o filme que queremos e a quantidade de "inspirações"

````
# Exemplo de uso da função retrieve_relevant_movies
# prompt = "A movie about space exploration"
prompt = "Um filme de um herói que ganhou super poderes e precisa salvar Nova Iorque"
````

````
# Traz 2 filmes semelhantes a query
num_to_retrieve = 2
````

````
overview_keys = list(overviews.keys())
````

### Gerando um resumo novo similar aos resumos anteriores:

````
template_generation = ChatPromptTemplate.from_template(f'''
    Você é um criador de sumários de filmes, seguindo a formatação e tamanho dos exemplos:
    EXEMPLOS:
    1: {overviews[overview_keys[0]]}

    2: {overviews[overview_keys[1]]}

    3: {overviews[overview_keys[2]]}

    4: {overviews[overview_keys[3]]}

    5: {overviews[overview_keys[4]]}

    - O filme não deve ser o sumário de um filme real
    - Não retorne nome de personagens
    Pergunta: {{entrada}}
''')

template_translate = ChatPromptTemplate.from_template(f'''
    You are a multilingual translator. Translate only to english. Return only the translation

    Input: {{entrada}}
''')
````

````
llm = ChatGroq(
    model="llama3-8b-8192",
    temperature= 1,
    max_retries=2
)
````

aplicando template para obter o resumo do filme

````
# Criar a primeira cadeia: Geração de sumário
chain_generation = LLMChain(
    llm=llm,
    prompt=template_generation
)

chain_translation = LLMChain(
    llm=llm,
    prompt=template_translate
)

summary = chain_generation.invoke({"entrada" : prompt})['text']
summary = chain_translation.invoke({"entrada" : summary})['text']
summary
````

Coletando o embedding do sumário, e buscando os outros filmes mais similares da base

````
relevant_movies = retrieve_relevant_movies(sentence_encoder, vector_DB_client, summary, num_to_retrieve)
# Exibir os filmes relevantes
print("Filmes relevantes:")
lista_descricao = []
lista_filme = []
for movie in relevant_movies:
    print(f"Filme: {movie['movie_name']}\nOverview: {movie['overview']}\n")
    texto = movie['overview']
    filme = movie['movie_name']
    lista_descricao.append(texto)
    lista_filme.append(filme)

````

````
print(lista_filme)
````

# Criar introdução de roteiro, baseado nos 3 filmes retornados

Utilizamos apenas a parte inicial dos roteiros

````
percentual_introducao = 1/4
````

Aplicando split na parte inicial dos filmes, e inserindo em nossa base de documentos para RAG

````
text_splitter = TokenTextSplitter(
    chunk_size=2000,
    chunk_overlap=20,
)

documents = []

for filme in lista_filme:
    try:
        file_name = metadatas[filme]['file']['file_name']
        with open(f'Movie-Script-Database/scripts/filtered/{file_name}.txt', 'r') as f:
            text = f.read()
        print(filme, "tem tamanho: ", len(text))

        idx_fim = int(len(text) * percentual_introducao)
        print("Introducão tem tamanho: ", idx_fim)

        textos = text_splitter.split_text(text[:idx_fim])
        for texto in textos:
            documents.append(texto)

    except Exception as e:
        print("Error: ", filme, e)
````

Criando nosso RAG

````
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-mpnet-base-v2")
````

````
vector_store = FAISS.from_texts(documents, embeddings)
vector_store
````

````
rag_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=vector_store.as_retriever()
)
````

# Criando introdução do filme com RAG

````
prompt = "Crie uma introdução para um filme, inspirado na introdução dos filmes disponíveis nos documentos. Retorne apenas o texto. A formatação deve ser a mesma dos documentos. A introdução deve ter pelo menos 3000 caracteres"

introducao = rag_chain.run(prompt)
````

````
introducao
````

````
from IPython.display import display, Markdown, Latex
Markdown(introducao)
````

# Baseado na introdução e no modelo com RAG, gerar continuamente um desenvolvimento e conclusão para o filme

````
documents = [introducao]
````

Iterativamente, criamos um RAG com as partes criadas do texto (iniciando somente com a introdução), e aumentamos a história gradativamente

````
for i in tqdm(range(10)):
    vector_store = FAISS.from_texts(documents, embeddings)
    rag_chain = RetrievalQA.from_chain_type(
        llm=llm,
        chain_type="stuff",
        retriever=vector_store.as_retriever()
    )
    desenvolvimento = rag_chain.run("Create a continuation for the story in the documents, the continuation must have at least 3000 characters and must keep the original movie-script formatting.")
    documents.append(desenvolvimento)
    time.sleep(60) # Esperando 1 minuto para chamada de API não exaurir
documents
````

Realizamos o mesmo processo, mas para criar um fim à história (mudando o prompt)

````
vector_store = FAISS.from_texts(documents, embeddings)
rag_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=vector_store.as_retriever()
)
desenvolvimento = rag_chain.run("Create an ending for the documents story. The end must have at least 5000 characters and must keep the original movie-script formatting.")
documents.append(desenvolvimento)
````

````
texto_final = "\n".join(documents)
Markdown(texto_final)
````
